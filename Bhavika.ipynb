```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
root = tk.Tk()
root.title("Student Management System")
root.geometry("600x450")
name_var = tk.StringVar()
roll_var = tk.StringVar()
marks_var = tk.StringVar()
search_var = tk.StringVar()
students = []
def add_student():
    name = name_var.get()
    roll = roll_var.get()
    marks = marks_var.get()
    if name == "" or roll == "" or marks == "":
        messagebox.showwarning("Input Error", "All fields are required")
        return
    student = (name, roll, marks)
    students.append(student)
    tree.insert("", tk.END, values=student)
    messagebox.showinfo("Success", "Student added successfully!")
    name_var.set("")
    roll_var.set("")
    marks_var.set("")
def search_student():
    search_name = search_var.get().strip().lower()
    found = False
    for child in tree.get_children():
        values = tree.item(child)["values"]
        name = values[0].lower()
        if search_name in name:
            tree.selection_set(child)
            tree.focus(child)
            tree.see(child)
            found = True
        else:
            tree.selection_remove(child)
    if not found:
        messagebox.showinfo("Not Found", "No student found with that name.")
tk.Label(root, text="Name").grid(row=0, column=0, padx=10, pady=10, sticky="w")
tk.Entry(root, textvariable=name_var).grid(row=0, column=1)
tk.Label(root, text="Roll No").grid(row=1, column=0, padx=10, pady=10, sticky="w")
tk.Entry(root, textvariable=roll_var).grid(row=1, column=1)
tk.Label(root, text="Marks").grid(row=2, column=0, padx=10, pady=10, sticky="w")

tk.Entry(root, textvariable=marks_var).grid(row=2, column=1)
tk.Button(root, text="Add Student", command=add_student).grid(row=3, column=1, pady=10)
tk.Label(root, text="Search by Name").grid(row=4, column=0, padx=10, pady=10, sticky="w")
tk.Entry(root, textvariable=search_var).grid(row=4, column=1)

tk.Button(root, text="Search", command=search_student).grid(row=4, column=2)
columns = ("Name", "Roll No", "Marks")
tree = ttk.Treeview(root, columns=columns, show="headings")

for col in columns:
    tree.heading(col, text=col)
    tree.column(col, width=150)
    
tree.grid(row=6, column=0, columnspan=3, padx=10, pady=20)
root.mainloop()
```
